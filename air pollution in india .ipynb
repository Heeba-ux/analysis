```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv("AirQuality.csv")
data.head(10)
```

```python
del data['lastupdate']
data.head()
```

```python
plt.plot(data['Avg'])
plt.xlabel('cities')
plt.ylabel('amount')
plt.title('Average Pollution Data')
```

```python
plt.plot(data['Max'])
plt.xlabel('cities')
plt.ylabel('amount')
plt.title('Maximum Pollution Data')
```

```python
plt.plot(data['Min'])
plt.xlabel('cities')
plt.ylabel('amount')
plt.title('Minimum Pollution Data')
```

```python
#checking which state have more pollution
plt.figure(figsize=(20,10), dpi = 80)
sns.countplot(x='State',data=data)
plt.xlabel('State')
plt.tight_layout()
```

```python
data_p1=data[data.Pollutants=='PM2.5']
data_p1[['Max','Avg','Min']].plot()
plt.xlabel('cities')
plt.ylabel('amount')
plt.title('PM2.5')
```

```python
#state wise
from pandas import DataFrame
df =DataFrame(data.State)
DataFrame.drop_duplicates(df)
```

```python
data_state1=data[data.State=='Andhra_Pradesh']
data_state1[['Min','Avg','Max']].plot()
plt.xlabel('cities')
plt.ylabel('Amount')
plt.title('Andhra_Pradesh')
```

```python
data_state7=data[data.State=='Karnataka']
data_state7[['Min','Avg','Max']].plot()
plt.xlabel('cities')
plt.ylabel('Amount')
plt.title('Karnataka')
```

```python
data_pollu=data.groupby('Pollutants')
data_pollu.mean()
plt.figure(figsize=(20,10) , dpi=100)
plt.plot(data_pollu.mean())
plt.legend(['Max','Avg','Mean'])
plt.xlabel('Amount')
plt.ylabel('Pollutant')
plt.title('Pollutant Amounts')
```

```python
#state pollution 
data_states=data.groupby('State')
data_states.mean()
```

```python
plt.figure(figsize=(18,5) , dpi=100)
plt.plot(data_states.mean())
plt.legend(['Max','Avg','Min'])
plt.xlabel('States')
plt.ylabel('Amount')
plt.tight_layout()
```

```python
data_city=data.groupby('city')
data_city.mean()
```

```python
plt.figure(figsize=(150,50) , dpi=100)
plt.plot(data_city.mean())
plt.xlabel('City')
plt.ylabel('Amount')
plt.show()
```
